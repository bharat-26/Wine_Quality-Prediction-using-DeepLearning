# Feature Engineering 

### Now, Here Random forest Regressor method is used for feature selection , like which feature is imp for the model 
#### Random Forest Regressor is used bcuz it gives the values in the form of gini impurity , info gain and entropy , Hence, this method is also used in feature engineering


```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(random_state = 1, max_depth = 12)
x = wine_df.drop(columns = 'quality', axis = 1)
           # OR 
# x = wine_df.iloc[:,:-1]
# print(x)
y = wine_df['quality']   
rf_model.fit(x,y)
features = wine_df.columns

# To get the importance of any feature ,one inbuild function : feature_importances_
importances = rf_model.feature_importances_  
print(importances)
print()

# to sort any numpy array , np.argsort() is there.... 
sorted_index = np.argsort(importances)[:]
print(sorted_index)


plt.figure(figsize = (16,10))
plt.title('Feature Importances', fontsize = 24)
plt.bar(range(len(sorted_index)),importances[sorted_index],color = 'blue')
# plt.yticks(range(len(sorted_index)),[features[i] for i in sorted_index], fontsize = 15)
plt.xlabel('Features_Importances',fontsize = 18)
plt.show()
```

```python
# Check For Multilinearity
plt.figure(figsize = (18,12))
sns.heatmap(x.corr(), annot = True)
plt.show()
```

# Handling Imbalanced Dataset 

#### Performing Oversampling And not undersampling, If we do Undersampling here then lots of data with larger imp have will get lost
##### Hence, Oversampling is performed

```python
import imblearn
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy = 'auto')  # u can also write "not majority"
x_sm,y_sm = smote.fit_resample(x,y) # sm = oversampling
```

```python
y_sm.value_counts()
```

###### Now, We are doing standardization to remove the biases 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_sm[x_sm.columns] = scaler.fit_transform(x_sm[x_sm.columns])
x_sm.head(8)
```

#### Dataset has been balanced

# Now , Model Selection

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_sm, y_sm, train_size = 0.8, test_size = 0.2,random_state = 42)
```

# Applying Artificial Neural Network

```python
import tensorflow 
from tensorflow.keras import utils 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout
```

```python
model = Sequential()
model.add(tensorflow.keras.layers.Input(shape = (11,)))
model.add(tensorflow.keras.layers.Dense(16, activation = 'relu'))
model.add(tensorflow.keras.layers.Dense(32, activation = 'relu'))
model.add(tensorflow.keras.layers.Dense(64, activation = 'relu'))
model.add(tensorflow.keras.layers.Dropout(0.3))
model.add(tensorflow.keras.layers.Dense(128, activation = 'relu'))
model.add(tensorflow.keras.layers.Dense(6, activation = 'softmax'))
```

```python
## 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.fit_transform(y_test)
```

```python
y_train
```

```python
y_train = pd.DataFrame(y_train)
y_test = pd.DataFrame(y_test)
```

```python
y_train
```

```python
y_test
```

```python
y_train  = tensorflow.keras.utils.to_categorical(y_train, 6)
y_test = tensorflow.keras.utils.to_categorical(y_test, 6)
```

```python
# print(y_train)
print()
print()
print(y_test)
```

```python
model.compile(optimizer = 'adam',loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x_train, y_train, epochs = 500, verbose = 1)
```
